22. Crie um programa em Python que imprima a soma dos n primeiros números primos, onde n é um número fornecido pelo usuário.

```
n = int(input('Digite um número inteiro positivo: '))
soma = 0
conta = 0
num = 2
while conta < n:
    primo = True
    for i in range(2, num):
        if num % i == 0:
           primo = False
           break
    if primo:
        soma += num
        conta += 1
    num += 1
print(f'A soma dos {n} primeiros números primos é: {soma}')
```

25. Crie um programa em Python que leia um número do usuário e verifique se ele é um número perfeito. Um número perfeito é aquele que é igual à soma de seus divisores próprios (ou seja, a soma de seus fatores inteiros positivos, excluindo ele próprio).

```
numero = int(input("Digite um número: "))
soma = 0

for i in range(1, numero):
    if numero % i == 0:
        soma += i

if numero == soma:
    print(f'O numero {numero} é perfeito. A soma dos seus divisores é: {soma}')
else:
    print(f'O número - {numero} - não é perfeito. A soma dos seus divisores é: {soma}')
```

26. Crie um programa em Python que calcule a série de Fibonacci para um determinado número de termos fornecido pelo usuário. A série de Fibonacci é uma sequência em que cada número é a soma dos dois números anteriores: 0, 1, 1, 2, 3, 5, 8, 13, ...

```
termos = int(input("Digite o número de termos desejados: "))
fibonacci = []
i = 1
n1 = 0
n2 = 1
n3 = 0

while i <= termos:
  fibonacci.append(n3)
  n1 = n2
  n2 = n3
  n3 = n1 + n2
  i += 1


print(f'Os primeiros {termos} termos de fibonacci são: {fibonacci}')
```

5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
arr = np.random.rand(4,4)

diagonal = np.diag(arr)
diagonal_inversa = np.fliplr(arr).diagonal()

media = np.mean([diagonal, diagonal_inversa])
desvio = np.std([diagonal, diagonal_inversa])

print(f'{arr}\n')

print(f'A diagonal principal é - {diagonal}.\n')
print(f'A diagonal inversa é - {diagonal_inversa}.\n')

print(f'Média diagonais: {media}\n')
print(f'Desvio padrão diagonais: {desvio}\n')
```
